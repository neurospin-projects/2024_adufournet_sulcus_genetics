## Check if the representation is an hypersphere

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
import os
import glob
```

```python
# Chose the framework to work with
path_to_ChampollionV0 = "/neurospin/dico/data/deep_folding/current/models/Champollion_V0"

# Select the region
region = "SC-sylv_left" #CINGULATE_right ORBITAL_left ORBITAL_right SC-sylv_left SC-sylv_right STi-STs-STpol_left STi-STs-STpol_right

path_to_region = os.path.join(path_to_ChampollionV0, region)

# Several models can exist for a same region
for date in glob.glob(path_to_region):

    # They are sorted by date
    for model in glob.glob(os.path.join(date, '*')):

        # Select the embeddings corresponding epoch 120
        path_to_embeddings = os.path.join(model, 'ukb_random_embeddings/full_embeddings.csv')

        if os.path.isfile(path_to_embeddings):
            embedding = pd.read_csv(path_to_embeddings, index_col=0)
            dist_from_origin = np.linalg.norm(embedding, axis=1)/embedding.shape[1]
            plt.hist(dist_from_origin, bins=80)
            plt.xlabel('Distance from origin')
            plt.show()

            

```
