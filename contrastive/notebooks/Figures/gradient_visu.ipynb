## To explain visually how to interpret the average from the regression

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cloud0 = 0.1*np.random.randn(100,2) + np.array([0.5,1.5])
cloud1 = 0.2*np.random.randn(100,2) + np.array([1.3,1.4])
cloud2 = 0.2*np.random.randn(100,2) + np.array([0.6,0.9])
cloud3 = 0.1*np.random.randn(100,2) + np.array([1.5,0.5])
```

```python
plt.figure(figsize=(36,24))
plt.scatter(cloud0[:,0], cloud0[:,1], color='steelblue')
plt.scatter(cloud1[:,0], cloud1[:,1], color='steelblue')
plt.scatter(cloud2[:,0], cloud2[:,1], color='steelblue')
plt.scatter(cloud3[:,0], cloud3[:,1], color='steelblue')
plt.axline((0.5,1.5), slope=-1, linewidth=4, color='black')
plt.savefig(f'/volatile/ad279118/Imaging_Genetics_2025/Figures_Report/Interpretation.eps', format='eps')
plt.show()
```

```python
plt.figure(figsize=(36, 24))

# Create a gradient background
x = np.linspace(0, 1, 500)
y = np.linspace(0, 1, 500)
X, Y = np.meshgrid(x, y)
gradient = X + Y  # Creates diagonal gradient (top left to bottom right)

# Display the gradient as background
plt.imshow(gradient, cmap="Reds", origin="upper", extent=[0, 2, 0, 2], alpha=0.5)

# Plot the scatter plots
plt.scatter(cloud0[:, 0], cloud0[:, 1], color='steelblue', s=100)  
plt.scatter(cloud1[:, 0], cloud1[:, 1], color='steelblue', s=100)
plt.scatter(cloud2[:, 0], cloud2[:, 1], color='steelblue', s=100)
plt.scatter(cloud3[:, 0], cloud3[:, 1], color='steelblue', s=100)

#plt.axline((0.5, 1.5), slope=-1, linewidth=4, color='black')
plt.show()
```
